# Exploratory Data Analysis - Olympics Dataset(1896-2016)

    In this notebook Exploratory Data Analysis is performed on Olympics Dataset (considered from 1896-2016)
    using python to analyze and visualize past olympics data and answer 
    specific questions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Loading the dataset
athlets=pd.read_csv('C:/Users/home/Desktop/New folder/Exploratory Data Analysis-Olympics Dataset/Dataset/athlete_events.csv')
region = pd.read_csv('C:/Users/home/Desktop/New folder/Exploratory Data Analysis-Olympics Dataset/Dataset/noc_regions.csv')
```

```python
athlets.head()
```

```python
region.head()
```

```python
#joining the dataframe using merge function(horizontal join)
athletes_df = athlets.merge(region,how = "left",on="NOC")
athletes_df
```

```python
athletes_df.shape
```

```python
#Formating the column name
athletes_df.rename(columns={'region':'Region','notes':'Notes'},inplace=True)
athletes_df
```

```python
athletes_df.info()
```

```python
athletes_df.describe()
```

```python
#checking the null values
nan_values=athletes_df.isna()
nan_columns=nan_values.any()
nan_columns
```

```python
athletes_df.isnull().sum()
```

```python
#printing the column names containing null values in the form of list
athletes_list=athletes_df.columns[athletes_df.isnull().any()].tolist()
athletes_list
```

```python
#Top 10 countries participating from 1896-2016
top_10_countries=athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
#plot for Top 10 countries participating from 1896-2016
plt.figure(figsize=(12,6))
plt.xticks(rotation=20)
plt.title("Overall participation by Country")
sns.barplot(x=top_10_countries.index,y=top_10_countries,palette="Set2")
```

```python
#Age distribution of the participants
plt.figure(figsize=(12,6))
plt.title("Age distribution of the participants")
plt.xlabel("Age")
plt.ylabel("Total participents")
plt.hist(athletes_df.Age,bins=np.arange(10,80,2),color="orange",edgecolor="white")
```

```python
winter_sports=athletes_df[athletes_df.Season=='Winter'].Sport.unique()
winter_sports
```

```python
summer_sports= athletes_df[athletes_df.Season=='Summer'].Sport.unique()
summer_sports
```

```python
#Male and Female Participents
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
#pie plot for male and female athlets
plt.figure(figsize=(12,6))
plt.title("Gender Distribution")
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%',startangle=150,shadow=True)
```

```python
#Total Medals
athletes_df.Medal.value_counts()
```

```python
#Total female athletes in each olympics
female_participants = athletes_df[(athletes_df.Sex=='F')&(athletes_df.Season=="Summer")][['Sex','Year']]
female_participants=female_participants.groupby('Year').count().reset_index()
female_participants .head(5)
```

```python
women_olympics =athletes_df[(athletes_df.Sex=='F')&(athletes_df.Season=="Summer")]
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(14,7))
sns.countplot(x='Year',data=women_olympics,palette="Spectral")
plt.title("Women participation")
```

```python
part=women_olympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(14,7))
part.loc[:,'F'].plot()
plt.title("Participation of female Athletes over time")
```

```python
#Gold Medal Athletes
gold_medals= athletes_df[(athletes_df.Medal=='Gold')]
gold_medals.head(5)
```

```python
#take only the values that are different from Nan
gold_medals = gold_medals[np.isfinite(gold_medals['Age'])]
```

```python
#Gold beyond 60
gold_medals['ID'][gold_medals['Age']>60].count()
```

```python
sporting_event = gold_medals['Sport'][gold_medals['Age']>60]
sporting_event
```

```python
#plot for sporting event
plt.figure(figsize=(10,5))
plt.tight_layout
sns.countplot(sporting_event)
plt.title("Gold Medal for athletes beyond 60 years")
```

```python
#Gold medals from each country
gold_medals.Region.value_counts().reset_index(name='Medal').head(5)
```

```python
totalGoldMedals=gold_medals.region.value_counts().reset_index(name='Medal').head(5)
g=sns.catplot(x='index',y='Medal',data=totalGoldMedals,height=5,kind='bar',palette='rocket')
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of medals")
plt.title("Gold Medals per country")
```

```python
#Rio olympics
max_year=athletes_df.Year.max()
max_year


team_names=athletes_df[(athletes_df.Year==max_year)&(athletes_df.Medal=="Gold")].Team
team_names.value_counts().head(10)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x= team_names.value_counts().head(20),y=team_names.value_counts().head(20).index)
plt.ylabel(None)
plt.xlabel("Country wise medals for the year 2016")
```

```python
#Medals won by Male and Female athletes according to their Height and weight
not_null_medals=athletes_df[(athletes_df['Height'].notnull())&(athletes_df['Weight'].notnull())]
```

```python
plt.figure(figsize=(12,10))
axis=sns.scatterplot(x='Height',y='Weight',data=not_null_medals,hue='Sex')
plt.title("Height v/s Weight of Olympic Medalist")
```
